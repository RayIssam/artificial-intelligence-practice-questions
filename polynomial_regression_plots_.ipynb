```python
# Polynomial Regression Plot with Alternative Style

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Use a different style that's commonly available
plt.style.use('ggplot')

# Function to generate data points
def make_data(N):
    np.random.seed(0)
    X = np.sort(np.random.rand(N))
    y = np.sin(2 * np.pi * X) + np.random.normal(scale=0.3, size=N)
    return X, y

# Generate 80 data points
X, y = make_data(80)
X = X[:, None]  # Reshape for compatibility with sklearn

# Test points for plotting smooth curves
X_test = np.linspace(-0.1, 1.1, 500)[:, None]

# Scatter plot of data points
plt.figure(figsize=(8, 6))
plt.scatter(X.ravel(), y, color='black', s=10, label="Data Points")
plt.xlim(-0.1, 1.0)
plt.ylim(-2, 12)

# Loop through specified polynomial degrees and plot
for degree in [1, 3, 5, 7, 9]:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)
    y_test = model.predict(X_test)
    plt.plot(X_test.ravel(), y_test, label=f'degree={degree}')

plt.legend(loc='best')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression for Degrees 1, 3, 5, 7, and 9")
plt.show()
```
