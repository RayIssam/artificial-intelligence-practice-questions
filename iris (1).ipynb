```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture

# Load the Iris dataset
iris_data = load_iris()
X_iris = iris_data.data
y_iris = iris_data.target

# Perform PCA to reduce the data to 2 components for visualization purposes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_iris)

# Create a DataFrame with PCA results and species information
iris_df = pd.DataFrame(X_pca, columns=["PCA1", "PCA2"])
iris_df['species'] = y_iris

# Define the number of clusters to try with GMM
cluster_counts = [2, 3, 4, 5]

# Set up the plotting area
fig, axes = plt.subplots(1, len(cluster_counts), figsize=(20, 5), sharey=True)

# Apply GMM with different cluster counts and plot the results
for i, n_clusters in enumerate(cluster_counts):
    # 1. Instantiate and fit the GMM model
    model = GaussianMixture(n_components=n_clusters, covariance_type='full', random_state=42)
    model.fit(X_iris)

    # 2. Predict cluster assignments
    iris_df['cluster'] = model.predict(X_iris)

    # 3. Plot the results for each cluster count
    sns.scatterplot(x="PCA1", y="PCA2", hue="cluster", data=iris_df, ax=axes[i], palette="viridis", s=60)
    axes[i].set_title(f"GMM with {n_clusters} Clusters")
    axes[i].set_xlabel("PCA1")
    axes[i].set_ylabel("PCA2")

plt.tight_layout()
plt.show()
```
